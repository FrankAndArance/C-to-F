```python
import pyautogui
from time import sleep, time
import keyboard
from IPython.display import clear_output
```

```python
'''
每秒判斷桌面上的圖片(座標)是否存在
'''

while True:
    sleep(0.1)
    
    #自動清除此格的文字輸出
    clear_output(wait = True)
    
    # 取得 Box 物件
    btn_0_location = pyautogui.locateOnScreen('images/BlueStacks.png',confidence = 0.9)
    
    print(btn_0_location)
    
    if keyboard.is_pressed('alt'):
        break
```
