```python
import pyautogui
from time import sleep, time
import keyboard
from IPython.display import clear_output
```

```python
'''
取得目前滑鼠座標
'''
while True:
    
     #自動清除此格的文字輸出
    clear_output(wait = True)
    
    #取得滑鼠游標的座標，並加以輸出
    x,y = pyautogui.position()
    print(f"x={x},y={y}")
    
    # 避免吃光記憶體
    sleep(0.1)
    
```

```python
'''
記錄滑鼠軌跡
'''

# 記錄軌跡用的變數
lisMove = []

def record():
    while True:
        
        # 按下 Ctrl 錄製軌跡，放開就不進行錄製
        if keyboard.is_pressed('ctrl'):
            x,y = pyautogui.position()
            lisMove.append(f"{x},{y}")
            sleep(0.1)
            
        # 按下 alt，結束程式
        elif keyboard.is_pressed('alt'):
            break
            
    with open('move.txt','w',encoding='utf-8') as file:
        file.write('\n'.join(lisMove))
        
'''
主程式
'''        
if __name__ =="__main__":
    record()
    
```

```python
'''
讀取滑鼠軌跡，並模擬滑鼠遊標移動
'''

def move():
    
     # 讀取軌跡檔
    with open('move.txt','r',encoding='utf-8') as file:
        
        # 逐行讀取
        for line in file:
            
             # 切割文字，找出 x,y
            x,y = line.strip().split(",")
            
            # 移動 (設定移動時間)
            pyautogui.moveTo(int(x),int(y),duration = 0.1)
            
            # 按alt結束
            if keyboard.is_pressed('alt'):
                break
                
if __name__ == "__main__":
    move()
            
```

```python
# 使用網址 https://html5.gamemonetize.co/7s9hk1ka1l8joyhwzkkn88dr4294bzam/

pyautogui.PAUSE = 0.05

# 打擊點
listPoint = [
    [220,800],
    [350,800],
    [480,800],
    [600,800]
]

def click():
    for list_ in listPoint:
        
        # 每次點擊都會重新取得指定打擊點的 RGB 值
        coord = pyautogui.pixel( list_[0], list_[1] )
        
        #coord[0]代表 Red，coord[1]代表 Green，coord[2]代表 Blue
        if coord[0] == 0 : 
            pyautogui.mouseDown( list_[0], list_[1] )
            pyautogui.mouseUp( list_[0], list_[1] )

            
if __name__ == "__main__":
    flag = False
    while True:
        
        # 按 alt 結束迴圈
        if keyboard.is_pressed('alt'):
            break
            
        # 按 ctrl後程式自動點擊
        if flag == False: 
            if keyboard.is_pressed('ctrl'):
                flag = True
                
        # 無限迴圈       
        if flag == True:
            click()
            
```
